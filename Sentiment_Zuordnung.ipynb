```python
from transformers import pipeline

texts = mini_df["text"].tolist()

print("Using BERT for Sentiment Analysis")
bert_model = pipeline("sentiment-analysis", model="nlptown/bert-base-multilingual-uncased-sentiment")
bert_predictions = bert_model(texts)

for text, prediction in zip(texts, bert_predictions):
    sentiment = prediction['label']
    confidence = prediction['score']
    print(f"Text: {text}\nSentiment: {sentiment} (Confidence: {confidence:.2f})\n{'-'*80}")
```

Model Description: This model is a fine-tune checkpoint of DistilBERT-base-uncased, fine-tuned on SST-2. This model reaches an accuracy of 91.3 on the dev set (for comparison, Bert bert-base-uncased version reaches an accuracy of 92.7).

Developed by: Hugging Face
Model Type: Text Classification
Language(s): English

```python
print("\nUsing DistilBERT for Sentiment Analysis")
distilbert_model = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
distilbert_predictions = distilbert_model(texts)

for text, prediction in zip(texts, distilbert_predictions):
    sentiment = prediction['label']
    confidence = prediction['score']
    print(f"Text: {text}\nSentiment: {sentiment} (Confidence: {confidence:.2f})\n{'-'*80}")
```

```python
from transformers import pipeline
mini_df= df.sample(n=100)
texts = mini_df["text"].tolist()
roberta_model = pipeline("sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment")

roberta_predictions = roberta_model(texts)
sentiments = []
confidences = []
label_mapping = {
    "LABEL_0": "negative",
    "LABEL_1": "neutral",
    "LABEL_2": "positive"
}

for prediction in roberta_predictions:
    sentiment = label_mapping[prediction['label']]
    confidence = prediction['score']
    sentiments.append(sentiment)
    confidences.append(confidence)

mini_df['sentiment'] = sentiments
mini_df['confidence'] = confidences
```
